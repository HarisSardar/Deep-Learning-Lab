### 1. Implement the step function as an activation function in Python using both the math library and TensorFlow. Test it with different input values and visualize its output using a simple plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import tensorflow as tf

# Step function using math library
def step_math(x):
    return 1 if x >= 0 else 0

# Step function using TensorFlow
def step_tf(x):
    return tf.cast(tf.greater_equal(x, 0), tf.float32)

# Test values
test_values = np.linspace(-5, 5, 100)

# Calculate outputs
outputs_math = [step_math(x) for x in test_values]

# Convert test values to TensorFlow tensor
test_values_tf = tf.constant(test_values, dtype=tf.float32)

# Calculate outputs using TensorFlow
outputs_tf = step_tf(test_values_tf)

# Plot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(test_values, outputs_math, label='Step (Math)', color='blue')
plt.title('Step Function (Math)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(test_values, outputs_tf, label='Step (TensorFlow)', color='red')
plt.title('Step Function (TensorFlow)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### 2. Create a Python function for the ReLU (Rectified Linear Unit) activation function using both the math library and TensorFlow. Test it with a range of input values and plot its output.

```python
from matplotlib import pyplot
import math
def Rectified_Linear_Unit(s):
    return max(0,s)
input = [s for s in range (-10,10)]
output = [Rectified_Linear_Unit(s) for s in input]

pyplot.plot(input,output)
pyplot.show()
```

```python
import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt

tf.disable_v2_behavior()

x = tf.placeholder(tf.float32, name='x')
relu_output = tf.nn.relu(x)  # Using tf.nn.relu instead of tf.tanh

with tf.Session() as sess:
    inputs = [i for i in range(-10, 10)]  # Fixed variable name and adjusted range

    result = sess.run(relu_output, feed_dict={x: inputs})  # Fixed session variable name

    print("ReLU of", inputs, "is", result)

plt.plot(inputs, result)
plt.grid()
plt.show()
```

### 3. Write a Python code to implement the Sigmoid activation function using both the math library and TensorFlow. Visualize its output and discuss its characteristics compared to other activation functions.

```python
from matplotlib import pyplot
import math

def sigmoid(s):
    return 1 / (11+ math.exp(-s))
input = [s for s in range (-10,10)]
output = [sigmoid(s) for s in input]

pyplot.plot(input,output)
pyplot.show()
```

### 4.	Develop a Python function for the Tanh (Hyperbolic Tangent) activation function using both the math library and TensorFlow. Test it with various input values and plot its output to observe its behavior.

```python
def Hyperbolic_Tangent(s):
    return ( math.exp(s) - math.exp(-s))/(math.exp(s) + math.exp(-s))
input = [s for s in range (-10,10)]
output = [Hyperbolic_Tangent(s) for s in input]

pyplot.plot(input,output)
pyplot.show()
```

### 5.	Implement the Leaky ReLU activation function in Python using both the math library and TensorFlow. Experiment with different leakage coefficients and plot the function's output.

```python
def Leaky_Rectified_Linear_Unit(s,a=0.1):
   
    return max(a*s,s)
input = [s for s in range (-10,10)]
output = [Leaky_Rectified_Linear_Unit(s) for s in input]

pyplot.plot(input,output)
pyplot.show()
```

```python
import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt

tf.disable_v2_behavior()

x = tf.placeholder(tf.float32, name='x')

# Define Leaky ReLU manually
def leaky_relu(x, alpha=0.01):
    return tf.maximum(alpha * x, x)

leaky_relu_output = leaky_relu(x)

with tf.Session() as sess:
    inputs = [i for i in range(-10, 10)]  # Fixed variable name and adjusted range

    result = sess.run(leaky_relu_output, feed_dict={x: inputs})  # Fixed session variable name

    print("Leaky ReLU of", inputs, "is", result)

plt.plot(inputs, result)
plt.grid()
plt.show()
```

### 6.	Create a Python function for the Softmax activation function using both the math library and TensorFlow

```python
def softmax(s):
    e_x = [math.exp(val) for val in s]
    sum_e_x = sum(e_x)
    return [i / sum_e_x for i in e_x]
input = [s for s in range (-10,10)]
output = softmax(input)

pyplot.plot(input,output)
pyplot.show()
```

```python
import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt

tf.disable_v2_behavior()

x = tf.placeholder(tf.float32, name='x')

# Define softmax function manually using TensorFlow operations
def softmax(x):
    e_x = tf.exp(x - tf.reduce_max(x))  # Subtracting the maximum value for numerical stability
    return e_x / tf.reduce_sum(e_x, axis=0)

# Apply softmax to the input
softmax_output = softmax(x)

with tf.Session() as sess:
    inputs = [i for i in range(-10, 10)]  # Fixed variable name and adjusted range

    result = sess.run(softmax_output, feed_dict={x: inputs})  # Fixed session variable name

    print("Softmax of", inputs, "is", result)

plt.plot(inputs, result)
plt.title('Softmax Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```
